```python
import pandas as pd
df=pd.read_csv("housing.csv")
df.head()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#preprossing the data
label_encode=LabelEncoder()
df["ocean_proximity_encoded"]=label_encode.fit_transform(df.ocean_proximity) #changing the categorical data int numerical
df=df.drop("ocean_proximity",axis=1)
df.ocean_proximity_encoded.value_counts()
```

```python
df.isnull().any()
```

```python
#filling the null values
df=df.fillna(df.mean())
df.isnull().any()

x=df.drop("median_house_value",axis=1)     #defining independent features
y=df.median_house_value                    #defining dependent feature

from sklearn.preprocessing import StandardScaler
data_scalar=StandardScaler()               #standard scalar for feature scaling
X=data_scalar.fit_transform(x)
print(X)

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)     #splitting data
print(x_train.shape)
print(x_test.shape)
```

```python
#model creation and fitting the data

lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
#testing the model with test data

y_test_pred=lr_model.predict(x_test)
mse_test_before=mean_squared_error(y_test,y_test_pred,squared=False)
rmse=np.sqrt(mse_test_before)
print("rmse test score",rmse)
```

```python
# applying polynomial regression 

from sklearn.preprocessing import PolynomialFeatures

degree=2
poly_features=PolynomialFeatures(degree=degree)      #applying poly features of degree 2
X_poly=poly_features.fit_transform(X)

X_train,X_test,Y_train,Y_test=train_test_split(X_poly,y,test_size=0.1,random_state=42)
lr_model_poly=LinearRegression()
lr_model_poly.fit(X_train,Y_train)

Y_test_pred=lr_model_poly.predict(X_test)
mse_test=mean_squared_error(Y_test,Y_test_pred,squared=False)
print("rmse:")
np.sqrt(mse_test)
```

```python
#changing degree for polynomial regression by using PCA

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
data_scalar=StandardScaler()
X=data_scalar.fit_transform(x)

pca=PCA(n_components=8)       #degree 8
X_pca=pca.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.1,random_state=42)

lr_model_poly=LinearRegression()
lr_model_poly.fit(X_train,y_train)

#testing the data

Y_test_pred=lr_model_poly.predict(X_test)
mse_test_after=mean_squared_error(y_test,Y_test_pred,squared=False)
r2_test_after=r2_score(y_test,Y_test_pred)
print(r2_test_after)
np.sqrt(mse_test_after)
```
